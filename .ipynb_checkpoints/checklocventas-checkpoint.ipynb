```python
import pandas as pd
import numpy as np
pelis = pd.read_csv('movies.csv')
pelis.head()


pelis_numericas = pelis.select_dtypes([int, float]).fillna("0")

pelis_numericas

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = pelis_numericas.drop("ventas", axis=1)

y = pelis_numericas.ventas
X_train, X_test ,y_train,  y_test = train_test_split(X, y)
regr = LinearRegression()

regr.fit(X_train , y_train)
predicciones = regr.predict(X_test)
```

```python
X_train.reset_index()
y_test.reset_index()
y_train.reset_index()
```

```python
train = X_train.join(y_train)
test = X_test.join(y_test)
test= test.reset_index()
test['pred_ventas'] = pd.DataFrame(predicciones)
test[:10]
```

```python
listdf = [train,test]
full_df = pd.concat(listdf)
full_df.drop(['index'], axis=1)
full_df.head()
```
